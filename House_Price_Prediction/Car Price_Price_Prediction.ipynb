```python
def carvalue(text):
    if text == "jaguar" or text == "buick" or text =="porsche" or text =="bmw":
        return "Luxury"
    else:
        if text == "audi" or text == "volvo":
            return "High Class"
        if text == "alfa-romero" or text== "mercury" or text =="peugeot" or text =="saab":
            return "Middle Class"
        if text =="chevrolet":
            return "extreme low"
        else:
            return "Low Class"
```

```python
dataset["CarCategory"] = dataset["CarBrand"].apply(carvalue)
```
